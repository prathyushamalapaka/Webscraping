```python
import time
from random import randint
from warnings import warn

import pandas as pd
import requests
from IPython.core.display import clear_output
from bs4 import BeautifulSoup
```

```python
# Create dictinary to store various aspects of IMDB movie data
names = []
genres = []
durations = []
years = []
ratings = []
metascores = []
votes = []
total_gross = []
urls = []

# Monitor time
request = 1
movie_no = 1

for i in range(0, 1):
    
    # Url
    url = "https://www.imdb.com/list/ls055592025/?sort=num_votes,desc&st_dt=&mode=detail&page=1&ref_=ttls_vm_dtl," \
          "desc&start={}&ref_=adv_nxt".format(page)

    # Send the request and Get the response
    response = requests.get(url)

    # Make Soup
    soup = BeautifulSoup(response.text, "html.parser")

    # Fetch the data
    lister_items = soup.find_all("div", class_="lister-item")

    # Get the data
    for single_item in lister_items:
        single_item_content = single_item.find("div", class_="lister-item-content")

        # Extract the data
        # Movie Name
        name = single_item_content.h3.a.get_text()
        
        #Genre
        try:
            genre = single_item_content.find("span", class_="genre").get_text().replace("\n", "").rstrip()
        except:
            pass

        # Movie Time
        duration = single_item_content.find("p", class_="text-muted").find("span", class_="runtime")
        if duration is not None:
            duration = duration.get_text()
        else:
            duration = "0"
        
        # Year the movie has been released
        year = single_item_content.h3.find("span", class_="lister-item-year").get_text()
        
        # Movie ratings by users
        rating = single_item_content.find("span", class_="ipl-rating-star__rating").get_text()
        
        # Metascore-> Average of critic ratings
        metascore = single_item_content.find("div", class_="ratings-metascore")
        if metascore is not None:
            metascore = metascore.span.get_text()
        else:
            metascore = "0"
            
        # Number of reviews/votes for the movie
        vote_gross = single_item_content.find("span",{"name": "nv"}).get_text()

        try:
            gross = single_item_content.find('span', text='Gross:').find_next('span')['data-value']
        except:
            pass
        
        #Obtaining the url for each movie page
        url = "https://www.imdb.com"+single_item_content.a.get('href')
        
        # Add data to global array
        names.append(name)
        genres.append(genre)
        durations.append(duration)
        years.append(year)
        ratings.append(rating)
        metascores.append(metascore)
        votes.append(vote_gross)
        total_gross.append(gross)
        urls.append(url)
        
        print("Movie No. #{}: {} added!".format(movie_no, name))

        # Separate Every Movie
        print("---------------------------------------------------")

        # Increment Movie
        movie_no += 1

    # If status is not 200 then show an warning with request number
    if response.status_code != 200:
        warn("{} no. has status code {}\n{}\n".format(request, response.status_code, url))

    # Increment Request
    request += 1

    # Sleep for sometime
    print("\n---------------------------------------------------")
    rand_sec = randint(1, 30)
    time.sleep(rand_sec)
    print("\tSleep For {}sec".format(rand_sec))
    print("---------------------------------------------------\n")
    clear_output(wait=True)

# Store the data to csv format
data_frame = pd.DataFrame({
    "Name": names,
    "Genre":genres,
    "Year": years,
    "Duration": durations,
    "User Ratings": ratings,
    "Critic Score": metascores,
    "Vote": votes,
    "Gross": total_gross,
    'URL':urls
})
```
