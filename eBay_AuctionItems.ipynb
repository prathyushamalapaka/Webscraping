```python
#Getting the page into a variable and creating a new file
page_lgphone = requests.get(url = lg_phone, headers = header) 

with open("ebay_lg_phone_1.htm", 'wb') as fd:
            fd.write(page_lgphone.content)
```

```python
pg_no = list(range(1,11))
webpages_lg = list(range(1,11))

#Appending the auction flag and page number.
for i in range(0,10):

    url2 =f"{lg_phone}&LH_Auction=1&_pgn={i+1}"
    lgphone_page2 = requests.get(url = url2, headers = header)
    
    #Storing each page in a list
    webpages_lg[i]= lgphone_page2 

    with open(f"lg_phone{i+1}.html", 'wb') as ebay_html:
            ebay_html.write(page_lgphone.content)

    print(f"Page {i+1} scraped.")
    time.sleep(10)
```

```python
#Each item in the above list stores an output from the html file
for i in range(1,11):
    with open("lg_phone"+ str(i) + ".html", encoding="utf8") as file:
        contents = file.read()
        
    #Parsing the html
    deals_parsed = BeautifulSoup(contents,'lxml')
    ab= deals_parsed.findAll("li", {"class" : "s-item s-item__pl-on-bottom s-item--watch-at-corner"})
    for l in ab:
        print(f"Product name : {l.find_all('h3')[0].text}")
        print(l.findAll('span',{ "class":"s-item__bids s-item__bidCount"})[0].text)
        print(f"Link : {l.a.get('href')}")
        print("\n")
    
```
